**Variables**

```python
# 1. Define the Cramer's V function
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    # Correcting for bias in small samples
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# 2. Select the categorical features you want to test
# Based on your notebook, these are the most interesting ones:
cat_features = ['Education', 'Marital Status', 'LoyaltyStatus', 
                'EnrollmentType', 'Income_Class', 'Active_Client', 'CLV_Tier']

# 3. Calculate the matrix
rows = []
for var1 in cat_features:
    col = []
    for var2 in cat_features:
        value = cramers_v(df_customer[var1], df_customer[var2])
        col.append(value)
    rows.append(col)

cramers_results = pd.DataFrame(rows, columns=cat_features, index=cat_features)

# 4. Plot the results
plt.figure(figsize=(10, 8))
sns.heatmap(cramers_results, annot=True, cmap='BuGn', fmt='.2f')
plt.title("Cramer's V: Association Strength Between Categorical Features")
plt.show()
```
